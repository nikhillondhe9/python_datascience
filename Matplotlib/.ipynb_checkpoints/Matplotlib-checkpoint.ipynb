```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure() #Create figure object fig
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_title('Plot 1')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
```

```python
ax1.plot(x,y)
ax2.plot(x,y)
fig
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
```

```python
ax.plot(x,z)
ax2.plot(x,y)
fig
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(8,2))
plt.tight_layout() 
```

```python
axes[0].plot(x,y,lw=3,ls='--',color='red')
axes[1].plot(x,z,lw=2,ls='-.',color='blue')
fig
```
